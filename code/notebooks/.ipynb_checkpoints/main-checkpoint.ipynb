```python
# Import all packages 
import numpy as np
import pandas as pd 
import pickle 
import seaborn as sns 
import itertools
import matplotlib.pylab as plt
from sklearn import tree
import graphviz
from sklearn.preprocessing import MinMaxScaler
```

```python
# Unpickling the scaler and the model 
scaler_file = '../models/scaler.pkl'
model_file = '../models/model.pkl'

infile_scaler = open(scaler_file, 'rb')
infile_model = open(model_file, 'rb')

new_scaler = pickle.load(infile_scaler)
new_model = pickle.load(infile_model)

infile_scaler.close()
infile_model.close()
```

```python
print ("Loaded scaler :: ", new_scaler)
print ("Loaded Decision tree model :: ", new_model)
```

```python
# Create synthetic data 
syth_data = []
for x in np.arange(-3, 2, 0.1):
    for y in np.arange(-3, 2, 0.1):
        syth_data.append((x,y))

df = pd.DataFrame(columns={'x', 'y', 'class'})
df['x'] = np.array(syth_data)[:,0]
df['y'] = np.array(syth_data)[:,1]
```

```python
y_hat = []

for i in range(len(df)):
    input_array = df[['x', 'y']].iloc[i].values.reshape(1,2)
    scaled_input = new_scaler.transform(input_array)
    model_output = new_model.predict(scaled_input.reshape(1,2))
    y_hat.append(model_output[0])

df['class'] = y_hat
```

```python
model_output
```

```python
df['class'].unique()
```

```python
fig = plt.figure(figsize = (10, 10))
sns.scatterplot(data = df, x = 'x', y ='y', hue ='class', palette='tab10')
plt.show()
```
